# A very simple Neural Network
In this notebook we will do some test with a very simple network. It mimics the XOR function.

````python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt

# XOR input and output
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]], dtype=np.float32)

y = np.array([[0],
              [1],
              [1],
              [0]], dtype=np.float32)

# Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2, input_dim=2, activation='sigmoid'),  # hidden layer
    tf.keras.layers.Dense(1, activation=None)                    # output layer without activation
])

# Compile the model with MSE loss
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
              loss='mse',
              metrics=['mse'])

# Train the model
history = model.fit(X, y, epochs=400, verbose=1)

# Plot learning curve
plt.plot(history.history['loss'])
plt.title('Learning Curve (MSE Loss)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

# Test the model
predictions = model.predict(X)
print("Predictions (raw outputs):")
print(predictions)

# Print learned weights
for i, layer in enumerate(model.layers):
    weights, biases = layer.get_weights()
    print(f"\nLayer {i+1} weights:\n{weights}")
    print(f"Layer {i+1} biases:\n{biases}")
````
